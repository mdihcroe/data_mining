```python
# of course I could write simple recursion without matrices, but I've written this instead 'cause I've studied this algorithm recently, so why not to use it?
''' Excercise #0.4 from Dasagupta book
	Fibonacci numbers counted by multiplying matrices:
					n
	F_n-1     0  1			F_0
	       =              *   
	F_n       1  1			F_1
'''

def Fib(n):
	matrix = np.array([[0, 1], [1, 1]])#array to be raised to the power of n-1
	matrix_powered = np.linalg.matrix_power(matrix, n-1)#raising to the power of n-1
	matrix_final = np.dot(matrix_powered, np.array([[0], [1]]))#consists of Fibb(n-1) and Fibb(n)	
	if n == 0: 
		return 0
	elif n == 1: 
		return 1
	else: 
		return matrix_final[1][0]#the bottom element of matrix_final

print (Fib(100))
```

```python
a, b, num = 100, 199, 11

def Linspace_Cycle(a, b, num):
    result = []
    for i in np.arange(a, b+1, (b-a)/(num-1)): #range() returns int and wouldn't work correctly with range in case of float step
        result.append(round(i, 1))
    return result

def Linspace_List_Comprehension(a, b, num):
        result = [round(i, 1) for i in np.arange(a, b+1, (b-a)/(num-1))]
        #result.append(b)
        return result

start = time.clock()
print ('np.linspace result: ', np.linspace (a, b, num), '\ntime: ', time.clock() - start, 'sec')
start = time.clock()
print ('cycle-based function result: ', Linspace_Cycle(a, b, num), '\ntime: ', time.clock() - start, 'sec')
start = time.clock()
print ('list comprehension used function result: ', Linspace_List_Comprehension(a, b, num), '\ntime: ', time.clock() - start, 'sec')
```

## Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline

def linear(t):
    return t/5 - 1

def cubic(t):
    return -14*t**3

def graph1(func_1, func_2, func_3, x_range):  
    x = np.array(x_range)  
    plt.title(r"Elementary functions", fontsize=20, color='#000624')
    plt.ylabel(r'ordinate', fontsize =16, color='#000624')
    plt.xlabel(r'abscissa', fontsize =16, color='#000624')
    plt.grid(True) 
    plt.plot(x, func_1(x), color = '#19008B', label=r'$y = sinx$') 
    plt.plot(x, func_2(x), color = '#C2DC50', label=r'$y = log_{10}x$') 
    plt.plot(x, func_3(x), color = '#FF2400', label=r'$y = \frac{x}{5} - 1$') 
    plt.legend(loc='best', fontsize=20).get_frame().set_alpha(0.5)
    
def graph2(func_1, func_2, x_range):  
    x = np.array(x_range)  
    plt.title(r"Elementary functions", fontsize=20, color='#000624')
    plt.ylabel(r'ordinate', fontsize =16, color='#000624')
    plt.xlabel(r'abscissa', fontsize =16, color='#000624')
    plt.grid(True) 
    plt.plot(x, func_1(x), color = '#91CB86', label=r'$y = shx$') 
    plt.plot(x, func_2(x), color = '#70377C', label=r'$y = -14x^3$')
    plt.legend(loc='lower left', fontsize=20).get_frame().set_alpha(0.5)
      
plt.figure(num=None, figsize=(16, 8), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(1, 2, 1)
graph1(np.sin, np.log10, linear, np.arange(0.1, 10, 0.1))
plt.subplot(1, 2, 2)
graph2(np.sinh, cubic, np.arange(0.1, 10, 0.1))
plt.tight_layout()
plt.savefig('elementary_func.png', bbox_inches='tight')
plt.show()
```
